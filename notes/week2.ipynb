```python
def main():
    number = get_number()
    meow(number)

def get_number():
    while True:
        n = int(input("What is n? "))
        if n > 0:
            break
    return n
def meow(n):
    for _ in range(n):
        print("meow")

main()
```

```python
#report

def main():
    s = {"name": "Voyager 1"}
    s.update({"distance": 163, "Orbit": "Sun"})
    print(create_report(s))

def create_report(s):
    return f"""
name: {s["name"]}
DIstance: {s.get("distance", "Unknown")}
orbit: {s.get("Orbit", "Unknown")}

"""

main()
```

```python
dict = {
    "A": 1,
    "B": 2,
    "C": 3
}

def main():
    for n in dict.keys():
        print(f"{n}: {dict[n]}: {convert(dict[n])}")

def convert(i):
    return i**2
main()
```

```python
words = "Hello World hhHhhhhh"
l_w = [word.lower() for word in words.split() if len(word) > 3]

dict = {word: l_w.index(word) for word in l_w}
print(f"{words}: {l_w}: {dict}")
```
