```python
pokemon_list = []

with open("s.csv") as file:
    for line in file:
        pok, type = line.rstrip().split(",")
        pokemon={"pok": pok, "type": type}
        pokemon_list.append(pokemon)

for p in sorted(pokemon_list, key = lambda pokemon: pokemon['pok']):
    print(f"{p['pok']}! {p['type']}!")
```

```python
import csv

pokemon_list = []

with open("s.csv") as file:
    r = csv.DictReader(file)
    for row in r:
        pokemon_list.append({"name": row["Name"], "attack": row['Attack']})

for p in sorted(pokemon_list, key=lambda pokemon: pokemon["name"]):
    print(f"{p['name']}! {p['attack']}!")
```

```python
import csv 

name = input("What's your name?")
home = input("Where is your home?")

with open("s.csv", "a") as file:
    writer = csv.DictWriter(file, fieldnames = ["name", "home"])
    writer.writerow({"name": name, "home": home})
```
