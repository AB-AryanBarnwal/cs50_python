```python
email = input("What's you email?").strip()

username, domain = email.split("@")
if (username) and domain.endswith(".edu"):
    print("Valid")
else:
    print("Invalid")
```

```python
import re

email = input("What's your email").strip()

if re.search(r"^(\w|\.)+@(\w+\.)?\w+\.(com|edu|in|gov|ord)$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")
```

```python
import re
name = input("What's your name?").strip()

if matches := re.search(r"^(.+), *(.+)$", name):
    name = matches.group(2) + " " + matches.group(1)
print(f"Hello {name}")
```

```python
import re

url = input("URL: ").strip()

username = re.sub(r"^((https?://)?(www\.)?)?twitter\.com/", "", url)
print(f"Username: {username}")
```

```python
import re

url = input("URL: ").strip()

if matches:=re.search(r"^https?://(?:www\.)?twitter\.com/([a-z0-9_]+)", url, re.IGNORECASE):
    print("Username:", matches.group(1))
```

```python
import re

def main():
    code = input("HEX code:")

    pattern = r"^#[0-6A-Fa-f]{6}$"
    match = re.search(pattern, code)
    if match:
        print(f"Valid. Matched with {match.group()}")
    else:
        print("Invalid")


main()
```

```python
import re

locations = {"+1":"United States and Canada","+62":"Indonesia","+505":"Nicaragua"}

def main():
    pattern = r"(?P<code>\+\d{1,3}) \d{3}-\d{3}-\d{4}"
    number = input("Number: ")

    match = re.search(pattern,number)
    if match:
        c = match.group("code")

        print(locations[c])
    else:
        print("Invalid")


main()
```
