```python
import random

cards = ["jack", "queen", "king"]

def main():
    print(random.sample(cards, k=2))

main()
```

```python
import random

cards = ["jack", "queen", "king"]

def main():
    print(random.choices(cards, k=2))

main()
```

```python
import random

cards = ["jack", "queen", "king"]

def main():
    print(random.choices(cards, weights = [1, 33, 63], k=2))

main()
```

```python
import random

cards = ["jack", "queen", "king"]

def main():
    random.seed(1)
    print(random.choices(cards, k=2))

main()
```
