```python
#compare
x = int(input("What's x? "))
y = int(input("What's y? "))

if x<y:
    print("x < y")

elif x>y:
    print("x > y")

elif x==y:
    print("x = y")

else:
    print("Sorry! We don't know.")
```

```python
#or
x = int(input("What's x? "))
y = int(input("What's y? "))

if x!=y:
    print("x not= y")

else:
    print("x = y")
```
