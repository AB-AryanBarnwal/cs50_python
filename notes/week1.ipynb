```python
#compare
x = int(input("What's x? "))
y = int(input("What's y? "))

if x<y:
    print("x < y")

elif x>y:
    print("x > y")

elif x==y:
    print("x = y")

else:
    print("Sorry! We don't know.")
```

```python
#or
x = int(input("What's x? "))
y = int(input("What's y? "))

if x!=y:
    print("x not= y")

else:
    print("x = y")

```

```python
#grade
score = int(input("Enter your score: "))

if 90 <= score <= 100:
    print("A")

elif 80 <= score:
    print("B")
elif 70 <= score:
    print("C")
elif 60 <= score:
    print("D")
else:
    print("F")


```

```python
#parity 
# Boolean

def main():
    x = int(input("What is x? "))
    if is_even(x):
        print("Even")
    else: 
        print("Odd")

def is_even(n):
    return n % 2 == 0
    
    
main()
```

```python
#House
#match

name = input("What is your name? ")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Griffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Who? ")
```
