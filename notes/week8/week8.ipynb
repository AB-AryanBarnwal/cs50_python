```python
class Vault:
    def __init__(self, galleons = 0, sickles=0, knuts=0):
        self.galleons = galleons
        self.knuts = knuts
        self.sickles = sickles
    
    def __str__(self):
        return f"{self.galleons} Galleons, {self.sickles} Sickles, {self.knuts} Knuts"
    
    def __add__(self, other):
        galleons = self.galleons + other.galleons
        sickles = self.sickles + other.sickles
        knuts = self.knuts + other.knuts
        return Vault(galleons, sickles, knuts)

potter = Vault(100, 50, 24)
weasley = Vault(25, 50, 100)
malfoy = Vault(100, 50, 25)
print(malfoy)
print(weasley)
print(potter)

total = potter + weasley + malfoy
print(total)
```

```python
class Package:
    def __init__(self, id, sender, reciever, weight):
        self.id = id
        self.sender = sender
        self.weight = weight
        self.reciever = reciever
    
    def __str__(self):
        return f"Package {self.id}: {self.sender} to {self.reciever}; weight: {self.weight} kg"
    
    def calculate_cost(self, cost_per_kg):
        return self.weight*cost_per_kg

def main():
    packages = [
        Package(id = 1, sender = "Alice", reciever="Bob", weight=10),
        Package(id = 2, sender = "Bob", reciever="Charlie", weight=1)
    ]

    for package in packages:
        print(f"{package} costs {package.calculate_cost(cost_per_kg=2)}")



main()
```

```python
class Food:
    base_heart = 1

    def __init__(self, ingredients):
        self.ingredients = ingredients
        self.heart = Food.calculate_hearts(ingredients)

    @classmethod
    def calculate_hearts(cls, ingredients):
        heart = cls.base_heart
        for ingredient in ingredients:
            if "hearty" in ingredient.lower():
                heart += 2
            else:
                heart += 1
        return heart
    
    @classmethod
    def from_nothing(cls, heart):
        food = cls(ingredients=[])
        food.heart = heart
        return food

def main():
    mushroom_skewer = Food(ingredients = ["Mushrooms", "Hearty Mushroom"])
    print(f"This skewer heals {mushroom_skewer.heart} hearts!")

    mushroom_skewer = Food.from_nothing(heart=2)
    print(f"This skewer heals {mushroom_skewer.heart} hearts!")


main()
```
